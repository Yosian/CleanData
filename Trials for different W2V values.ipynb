```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
```

```python
texto=pd.read_csv('/Users/empathybroker/Documents/Synonyms/cbn-data-experiment-synonyms/zaracleansearches.csv', sep=',', header=0)
texto.drop(['Unnamed: 0'],axis=1, inplace=True)
texto.dropna(inplace=True)
```

```python
#organise DataFrame by product_id to keep similar products close and prepare words
texto.set_index('pnum', inplace=True)
texto.sort_index(inplace=True)
```

```python
words1=[i.split() for i in texto['fsearch']]
words2=[j.split() for j in texto['ssearch']]
#now we stagger both into a single column so both types of searches are next to each other
#words=[None]*(len(words1)+len(words2))
#words[::2]=words1
#words[1::2]=words2
```

```python
sent=[]
for i in range(len(words1)):
    w=words1[i]+words2[i]
    sent.append(w)
```

```python
del texto
del words1
del words2
```

```python
zarawords=Word2Vec(sentences=sent, size=100, window=2, min_count=50, workers=4, sg=1)
#These parameters are to be changed depending on what we want to achieve
```

```python
#in the next two lines I decided to couple searches so I would obtain longer sentences in the 
#hope to imrpove accuracy
```

```python
indwords=[]
for couple in sent:
    for j in couple:
        indwords.append(j)
```

```python
synonyms=[]
wordset=[]
for i in set(indwords):
    try:
        w=zarawords.wv.most_similar(i,topn=3)
        wordset.append(i)
        synonyms.append(w)
    except:
        pass
synonymsdf=pd.DataFrame(columns=['word','synonyms'])
synonymsdf['word']=wordset
synonymsdf['synonyms']=synonyms
```
