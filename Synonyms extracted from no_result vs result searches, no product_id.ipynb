```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
fsearch=zara[zara.columns[0]].astype(str).values.tolist()#need column as list to vectorise
```

```python
ssearch=zara[zara.columns[1]].astype(str).values.tolist()
```

```python
#I create a single vector so the dictionary includes all words, I keep first searches
#in even positions and second searches in odd positions
totsearch=[None]*(len(fsearch)+len(ssearch))
totsearch[::2]=fsearch
totsearch[1::2]=ssearch
totsearch
```

```python
vectorizer=CountVectorizer()
#hence I have now a staggered list, now I can vectorise
vec_totsearch=vectorizer.fit_transform(totsearch)
```

```python
vectotsearch=vec_totsearch.toarray()
```

```python
vectotsearch #just checking...
```

```python
even=vectotsearch[0::2]
```

```python
odd=vectotsearch[1::2]
```

```python
compar_vec=even-odd #I create two vectors as I need them to operate with them and draw conclusions
```

```python
differ_vec=[abs(k) for k in compar_vec] #I convert all into positive values so when I add I have
#all weights and they don't cancel out
```

```python
comparison=vectorizer.inverse_transform(compar_vec)
```

```python
type(comparison)#just checking
```

```python
len(comparison)#Just checking it is as long as the original search
```

```python
zara
```

```python
evensum=np.sum(even,axis=1) #variable I may need later
```

```python
oddsum=np.sum(odd,axis=1) #variable I may need later
```

```python
weight=evensum+oddsum #This is because long searches are usually more complex, I need a variable
#to tell me how useful a search may be and hence how trustworthy the synonyms
```

```python
difference=np.sum(differ_vec,axis=1)
```

```python
comparador=pd.DataFrame({"Weights":weight,"Difference":difference,"Candidates":comparison})
```

```python
comparador #this dataframe tells me what words may be synonyms. If the difference is 1 is probably
#a broader search, if the weight is high the synonyms are usually not useful
# the best synonyms are at Difference=2 and Weight=2,4
```
